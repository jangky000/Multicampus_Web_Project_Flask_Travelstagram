```python
from bs4 import BeautifulSoup
import urllib.request as REQ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# 환경 설정
matplotlib.rcParams['font.family']='Malgun Gothic' # 제어판 글꼴 속성
matplotlib.rcParams['axes.unicode_minus'] = False # 유니코드를 적용해라
```

```python
# 맛집 데이터 분석
# https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss201
url_base ='https://www.menupan.com'
url_sub ='/restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss201'
response = REQ.urlopen(url_base + url_sub)
soup = BeautifulSoup(response, 'html.parser')
soup
```

```python
# 하나만 뽑아오기 => select_one()
# 속성을 가지고 올 때는 딕셔너리 형태로 키값을 줘서 가져온다
menuList = []
hrefList = []
ul = soup.select_one('.rankingList .list')
for li in ul.select('li'):
    href = li.select_one('.listName a')['href']
    hrefList.append( href )
    rank = li.select_one('.numTop, .rankNum').string # class의 '또는'을 표현할 때 ',' 로 구분****************
    listName = li.select_one('.listName').string
    listType = li.select_one('.listType').string
    listArea = li.select_one('.listArea').string
    menuList.append( {'랭킹':rank, '상호명':listName, '업종':listType, '지역':listArea } )
#     print('='*30)
```

```python
hrefList
```

```python
df = pd.DataFrame(menuList)
df
```

```python
telList= []
addList = []
url_base ='https://www.menupan.com'
for href in hrefList:
    response = REQ.urlopen(url_base + href)
    soup = BeautifulSoup(response, 'html.parser')
    telList.append( soup.select_one('.tel1').string )
    addList.append( soup.select_one('.add1').string )
    print(soup.select_one('.tel1').string )
    print(soup.select_one('.add1').string )
```

```python
df['전화번호'] = telList
df['주소'] = addList
df
```

```python
df.to_excel('menu.xlsx')
```
