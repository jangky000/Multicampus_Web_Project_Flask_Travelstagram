```python
from selenium import webdriver
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import time

import pandas as pd
import numpy as np

%matplotlib inline

class instagram_hashtags:
    def hashtag_crawling(self, keyword, path, number):
        self.keyword = keyword
        self.path = path
        self.number = number
        
        tag_data = []
        words = []
        words_list = []
        
        # 국가: 총 검색수 : 태그들
        row_list = []
        # row들
        total_list =[]
        
        # 크롬창을 띄우지 않는 옵션을 넣는다
#         options = webdriver.ChromeOptions()
#         options.add_argument('headless')
#         options.add_argument('disable-gpu')
#         driver = webdriver.Chrome( path, options=options)
        
        driver = webdriver.Chrome(path)
        
        driver.get('https://www.instagram.com/explore/tags/' + keyword)
        time.sleep(3)

        totalCount = driver.find_element_by_class_name('g47SY').text

#         time.sleep(1)
        driver.find_element_by_xpath("//div[@class = '_9AhH0']").click()

        totalCount = pd.Series(totalCount)
        totalCount.replace(',', '', regex=True, inplace=True)
    
        if number > int(totalCount):
            number = int(totalCount)
        print(number)

        for i in range(number-1):
            time.sleep(2)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')

            tag_data.append(soup.select('div.C4VMK > span > a'))

            driver.find_element_by_xpath("//a[@class = 'HBoOv coreSpriteRightPaginationArrow']").click()


        for x in range(number-1):
            for y in range(len(tag_data[x])):
                words += tag_data[x][y]

        for i in words:
            if i[0] == '#':
                words_list.append(i)
        #         
        words_list2 = '|'.join(words_list)
        row_list.append(keyword)
        row_list.append(totalCount)
        row_list.append(words_list2)
        
        
        return row_list
```
