```python
from selenium import webdriver
from bs4 import BeautifulSoup
# from wordcloud import WordCloud
import matplotlib.pyplot as plt
import time

import pandas as pd
import numpy as np

%matplotlib inline

class instagram_hashtags:
    def hashtag_crawling(self, keyword, path, number):
        self.keyword = keyword
        self.path = path
        self.number = number
        
        tag_data = []
        words = []
        words_list = []
        
        # 국가: 총 검색수 : 태그들
        row_list = []
        # row들
#         total_list =[]
        
        # 크롬창을 띄우지 않는 옵션을 넣는다
#         options = webdriver.ChromeOptions()
#         options.add_argument('headless')
#         options.add_argument('disable-gpu')
#         driver = webdriver.Chrome( path, options=options)
        
        driver = webdriver.Chrome(path)
        
        try:
            driver.get('https://www.instagram.com/explore/tags/' + keyword)
        except Exception as e:
            row_list.append(keyword)
            row_list.append('0')
            row_list.append('')
            return row_list
        else:
            pass
        
        time.sleep(3)

        try:
            totalCount = driver.find_element_by_class_name('g47SY').text
        except Exception as e:            
            row_list.append(keyword)
            row_list.append('1')
            row_list.append('직접 입력')
            return row_list
        else:
            pass
        
        try:
            driver.find_element_by_xpath("//div[@class = '_9AhH0']").click()
        except Exception as e:            
            row_list.append(keyword)
            row_list.append('0')
            row_list.append('자료 없음')
            return row_list
        else:
            pass
        
        totalCount = totalCount.replace(',','')
        
        if number > int(totalCount):
            number = int(totalCount)
        print(number)

        for i in range(number-1):
            time.sleep(2)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')

            tag_data.append(soup.select('div.C4VMK > span > a'))

            driver.find_element_by_xpath("//a[@class = 'HBoOv coreSpriteRightPaginationArrow']").click()


        for x in range(number-1):
            for y in range(len(tag_data[x])):
                words += tag_data[x][y]

        for i in words:
            if i[0] == '#':
                words_list.append(i)
        #         
        words_list2 = '|'.join(words_list)
        row_list.append(keyword)
        row_list.append(totalCount)
        row_list.append(words_list2)
        
        
        return row_list
                
    def save_file_wordcloud(self, data_path, words_list):
        self.data_path = data_path
        self.words_list = pd.Series(words_list)
        
#         f = open(data_path, 'w')
#         for i in range(len(words_list)):
#             f.write(words_list[i] + "\n")
#         f.close()

        self.words_list.to_excel('tag_words.xlsx')
        
#         text = open(data_path).read()
        
#         wordcloud = WordCloud(max_font_size = 30, font_path = '/C:\Windows\Fonts/malgunbd.ttf', background_color = 'white').generate(text)
        
#         return wordcloud
    
    def Visualization(self, wordcloud_data, size = (40, 40), interpolation = 'bilinear', data_save = 'test'):
        self.size = size
        self.wordcloud_data = wordcloud_data
        self.interpolation = interpolation
        self.data_save = data_save
        
        plt.figure(figsize = size)
        plt.imshow(wordcloud_data, interpolation = interpolation)
        plt.axis('off') # x, y축 scale(숫자) 없애기.
        plt.show()
        plt.savefig(data_save) # 이미지 파일 저장
```
